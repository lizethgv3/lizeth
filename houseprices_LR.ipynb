```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('HousePrices.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.shape
```

```python
y=df['Price']
```

```python
y.shape
```

```python
y
```

```python
# to define x it can be
# x=df[['column1','column2','etc']] or use drop function

x=df.drop('Price', axis=1)
```

```python
x.shape
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
x=x.drop('Date',axis=1)
x.info()
```

```python
x=mm.fit_transform(x)

# get x variable standarized 

# standarization is a common requirement for ML estimators implemented in scikit-learn. They might 
# behave badly if the individual features do not more or less look like normally distributed data

# next approach is to go for minmax scaler
```

```python
x
```

Get Train and Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

Get Model Train

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

Get Model Prediction

```python
y_pred=lr.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

Get Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test, y_pred)
```

Get Visualization of Actual vs. Predicted Results

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs. Predicted Price")
plt.show()
```
